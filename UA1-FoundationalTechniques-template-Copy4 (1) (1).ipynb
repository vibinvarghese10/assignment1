```python
import pandas as pd
```

```python
import plotly.express as px
```

```python
url = " https://www.data.qld.gov.au/dataset/db190f2d-f866-4811-9a6e-4b78744b551b/resource/0f97b985-f5c7-49d2-8b0a-bc5dfbe070b9/download/advance-queensland-funding-recipients.csv"
df=pd.read_csv(url, encoding='ISO-8859-1')
```

```python
df['Approval date'] = pd.to_datetime(df['Approval date'], dayfirst=True, errors='coerce')
df['Year'] = df['Approval date'].dt.year
```

```python
df['Actual Contractual Commitment ($)'] = pd.to_numeric(df['Actual Contractual Commitment ($)'].replace('[,$]', '', regex=True), errors='coerce')
```

```python
# Fill missing values in the DataFrame
df.fillna({
    'Physical Address of Recipient - Post Code': df['Physical Address of Recipient - Post Code'].median(),
    'University Collaborator (if applicable)': 'Unknown',
    'Other Partners; Collaborators (if applicable)': 'Unknown'}, inplace=True)
```

## 1. Total Funding Trend
Funding reached a peak in 2017 of around 61 miilion before decreasing. The peak may reflect a period of high investment in key initiatives, while the subsequent decline could be due to a variety of circumstances, such as the completion of big projects, a strategic shift in priorities, or budget cuts. The results of financed projects during peak years can be used to determine success.From a decrease after 2017 it moves to slight increase in 2020 of around 35 million dollars.
Here a line graph is used to describe the trend for funding over the time.Where the dataframe is grouped by year and the Actual Contractual Commitment sums up the values for each year.

```python
fig = px.line(df.groupby('Year')['Actual Contractual Commitment ($)'].sum().reset_index(), 
                  x='Year', y='Actual Contractual Commitment ($)',
                  labels={'Actual Contractual Commitment ($)': 'Total Funding ($)'}, 
                  title='Total Funding Over Time')
fig.show()
```

## 2.	Funding by Region:
South-East Queensland receives considerably more funds than other regions. This could indicate population density, a concentration of research institutions, or a policy focus. To determine success, we'd need to compare the regional funding against the objectives of the program and regional needs.From the bar graph it is clear that South-East Queensland get more fund of around 153 million dollars where regional areas is getting only around 67 million dollars.
This code effectively groups funding data into areas depending on whether  they are part of South-East Queensland or regional, and then visualizes the total funding amount for each region using a bar chart, allowing for simple comparison of funding allocation between regions.duing this visualising 

```python
regions = ['Sunshine Coast (R)', 'Somerset', 'Moreton Bay (R)', 'Lockyer Valley (R)',
                  'Brisbane (C)', 'Ipswich (C)', 'Redland (C)', 'Logan (C)', 'Scenic Rim (R)', 'Gold Coast (C)']

# Function to classify as 'South-East Queensland' or 'Regional Queensland'
def classify_region(council):
    if council in regions:
        return 'South-East Queensland'
    else:
        return 'Regional Queensland'

# Apply the classification function to the 'Local Government /Council' column
df['Region'] = df['Local Government /Council'].apply(classify_region)
pic = px.bar(df.groupby('Region')['Actual Contractual Commitment ($)'].sum().reset_index(),
                 x='Region', y='Actual Contractual Commitment ($)',
                 labels={'Actual Contractual Commitment ($)': 'Total Funding ($)'},
                 title='Funding by Region')
pic.show()
```

## 3.Maximum and Minimum Grants by Program: 
The maximum grant is significantly higher than the minimum grant.In this bar chart there is 2 grant options to show data inside the max grant and min grant.When comparing the max grant betwwen the regions there is huge differene is there that is max grand for south-east quuensland is around 10 million and for regional it is 3 million.while min grant have slight diffrence.   

```python
df['Region Category'] = df['Local Government /Council'].apply(classify_region)
# Calculate maximum and minimum grants for each region
max = df.groupby('Region')['Actual Contractual Commitment ($)'].max().reset_index()
min = df.groupby('Region')['Actual Contractual Commitment ($)'].min().reset_index()

# Merging max and min grants into a single DataFrame
max['Grant Type'] = 'Max Grant'
min['Grant Type'] = 'Min Grant'
grants = pd.concat([max, min], ignore_index=True)

# Plotting 
pic = px.bar(grants, x='Region', y='Actual Contractual Commitment ($)',
             color='Grant Type', barmode='group',
             title='Maximum and Minimum Grants by Region',
             labels={'Actual Contractual Commitment ($)': 'Grant Amount ($)'})

pic.show()
```

## 4.Top 10 Total Grant Amounts Approved for Each Program:
This graph provides insights into which areas are receiving the most funding. The spread among the top-funded programs can reflect strategic focuses of the government. The “ignite ideas fund” notably received the most funding, suggesting a focus on innovation and startups.

```python
df['Region Category'] = df['Local Government /Council'].apply(classify_region)
# Group by 'Region Category' and 'Program' and sum the 'Actual Contractual Commitment ($)' for each group
# Sort the results in descending order
grants_per_program_region = df.groupby(['Region Category', 'Program'])['Actual Contractual Commitment ($)'].sum().sort_values(ascending=False)
# Select the top 10 entries from the sorted list for detailed visualization
top_grants_per_program_region = grants_per_program_region.head(10).reset_index()
#creating bar graph
fig = px.bar(top_grants_per_program_region, x='Program', y='Actual Contractual Commitment ($)',
             color='Region Category',  
             title='Top 10 Total Grant Amounts Approved for Each Program by Region',
             labels={'Actual Contractual Comm|itment ($)': 'Total Grant Amount ($)'})
fig.update_layout(xaxis_title='Program',
                  yaxis_title='Total Grant Amount ($)',
                  xaxis_tickangle=-45, height=1000, width=1000)  # Rotate labels for better readability
#viewing te graph
fig.show()
```

## 5.Total Grant Amount by Cycle
The middle cycle (2016–2024) showed an increase in funding, maybe due to increased program activity or a government initiative. The success of this phase will depend on the outcomes and impacts realized during this period.

```python
df['Region Category'] = df['Local Government /Council'].apply(classify_region)
# Filter to only include South-East Queensland and Regional Queensland
filtered_df = df[df['Region Category'].isin(['South-East Queensland', 'Regional Queensland'])]

# Define cycles
cycles = {
    '2012-2016': (2012, 2016),
    '2016-2020': (2016, 2020),
    '2020-2024': (2020, 2024)
}
# Calculate total grants per cycle for the filtered regions
cycle_data = []
for cycle_name, (start_year, end_year) in cycles.items():
        total_grant = filtered_df[(filtered_df['Year'] >= start_year) & (filtered_df['Year'] <= end_year)]['Actual Contractual Commitment ($)'].sum()
        cycle_data.append({'Cycle': cycle_name, 'Total Grant Amount ($)': total_grant})

cycle_grant = pd.DataFrame(cycle_data)
# Create the line chart
graph = px.line(cycle_grant, x='Cycle', y='Total Grant Amount ($)',
                title='Total Grant Amount by Cycle for SEQ and Regional Queensland',
                markers=True, 
                labels={'Total Grant Amount ($)': 'Total Grant Amount ($)', 'Cycle': 'Grant Cycle'})

# Update layout
graph.update_layout(xaxis_title='Grant Cycle',
                    yaxis_title='Total Grant Amount ($)',
                    xaxis=dict(tickmode='linear'))  

graph.show()
```

## 6 Total Of Grant Amounts Approved for Each Program by Region:
This plot effectively combines and visualizes significant financial data, providing insights into how grant funding is distributed across various programs in the given regions. It highlights the total funding and the variability within that funding, assisting in decision-making and analysis for parties interested in regional funding allocations.By analysing this histogram it shows the ignite ideal fund in south east queensland most grand amount

```python
df['Region Category'] = df['Local Government /Council'].apply(classify_region)
# Filtering to include only the categorized regions if other regions are present
filtered = df[df['Region Category'].isin(['South-East Queensland', 'Regional Queensland'])]

# Calculate total grants per program within the filtered regions, then sort
grants = filtered.groupby(['Region Category', 'Program'])['Actual Contractual Commitment ($)'].sum().reset_index()

# Creating a histogram with an overlay of a box plot for the filtered data
amount_show = px.histogram(grants, x='Program', y='Actual Contractual Commitment ($)',
                         color='Region Category',  # Differentiate by region
                         title='Sum Of Grant Amounts Approved for Each Program by Region',
                         labels={'Actual Contractual Commitment ($)': 'Total Grant Amount ($)'},
                         barmode='overlay')

# Adding a box plot trace
box = px.box(grants, x='Program', y='Actual Contractual Commitment ($)')
amount_show.add_trace(box.data[0])

# Update layout
amount_show.update_layout(width=1000, height=1000)

# Show the plot
amount_show.show()
```

## 7.Trends in funding 
Determine whether sectors or programs have received more, less amount of cash. This may reflect a change in governmental priorities. Alignment with Government Policies: By observing the timing and pattern of these changes,we can determine whether funding changes match to known government policy announcements or movements in public priorities.it can be seen that ignite ideas fund got most funding in the year 2017.

```python
df['Region Category'] = df['Local Government /Council'].apply(classify_region)
filtered = df[df['Region Category'].isin(['South-East Queensland', 'Regional Queensland'])]

# Grouping the filtered data by Year and Program, then summing up the funding
funding_over_time = filtered.groupby(['Year', 'Program'])['Actual Contractual Commitment ($)'].sum().reset_index()

# Creating the line chart
program = px.line(funding_over_time, x='Year', y='Actual Contractual Commitment ($)', color='Program',
                      title='Trends in Funding Allocation Over Time by Program (South East Queensland and Regional Queensland)',
                      labels={'Actual Contractual Commitment ($)': 'Total Funding ($)', 'Program': 'Sector/Program'})

# Updating layout to ensure every year is marked
program.update_layout(xaxis={'dtick': 1})


program.show()
```
